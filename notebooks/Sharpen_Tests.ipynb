```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import os
```

```python
stacked_image_file = 'stacked_jupiter.jpg'
```

```python
img = cv2.imread(stacked_image_file)
```

```python
plt.imshow(img)
plt.show()
```

```python
lp_filter = cv2.Laplacian(img, cv2.CV_64F)
```

```python
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened_image = cv2.filter2D(img, -1, kernel)
plt.imshow(sharpened_image)
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
src = cv2.GaussianBlur(RGB_img, (3, 3), 0)
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
edges = cv2.Laplacian(gray, -1, ksize=3, scale=1,delta=0,borderType=cv2.BORDER_DEFAULT)
```

```python
plt.imshow(edges)
plt.show()
```

```python
alpha = .5
beta =( 1.0 - alpha )
weighted_image = cv2.addWeighted( gray, alpha, edges, beta, 0.0)
```

```python
plt.imshow(weighted_image)
plt.show()
```

```python
edges2 = cv2.Laplacian(RGB_img, -1, ksize=3, scale=1,delta=0,borderType=cv2.BORDER_DEFAULT)
```

```python
alpha = .9
beta =( 1.0 - alpha )
weighted_image = cv2.addWeighted(img, alpha, edges2, beta, 0.0)
plt.imshow(weighted_image)
plt.show()
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
lap = cv2.Laplacian(gray_img, cv2.CV_64F)
sharp_img = gray_img - .7*lap
plt.imshow(sharp_img, cmap='gray')
plt.show()
```

```python
img_B = img[:,:,0]
img_G = img[:,:,1]
img_R = img[:,:,2]
```

```python
lap_B = cv2.Laplacian(img_B, cv2.CV_64F)
lap_G = cv2.Laplacian(img_G, cv2.CV_64F)
lap_R = cv2.Laplacian(img_R, cv2.CV_64F)
```

```python
sharp_B = img_B - .7*lap_B
sharp_G = img_G - .7*lap_G
sharp_R = img_R - .7*lap_R
```

```python
sharp_img = np.zeros(np.shape(img))
sharp_img[:,:,0] = sharp_B
sharp_img[:,:,1] = sharp_G
sharp_img[:,:,2] = sharp_R
```

```python
sharp_img = sharp_img.astype(np.uint8)
plt.imshow(sharp_img)
plt.show()
```

```python
blurred = cv2.GaussianBlur(img, (3,3), 0)
lap1 = cv2.Laplacian(blurred, cv2.CV_64F)
sharp = img - (.7*lap1)
sharp = sharp.astype(np.uint8)
plt.imshow(sharp)
plt.show()
```
